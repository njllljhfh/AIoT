## 作业4
使用题目二中数据，开始学生们开始减肥
- 本科生减肥，减掉的体重统一是10
- 博士生减肥，减掉体重范围是5~10

```python
# 问题一
# 先找到本科生
cond = df['学位'] == '本科'
# display(cond)

print('本科生减肥前：')
display(df[cond])

print('本科生减肥后：')
df.loc[cond,'体重'] -= 10  # 改变原数据
display(df[cond])
```

```python
# 问题二
# 先找到博士生
cond = df['学位'] == '博士'
# print(cond)
# print(cond.sum())  # False = 0，True = 1

print('博士生减肥前：')
display(df[cond])

print('博士生减肥后：')
# df[cond].shape[0] 获取博士的个数
# df.loc[cond,'体重'] -= np.random.randint(5,11,size=df[cond].shape[0])
df.loc[cond,'体重'] -= np.random.randint(5,11,size = cond.sum())
display(df[cond])
```
